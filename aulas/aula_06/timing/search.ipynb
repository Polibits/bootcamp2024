```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
linear_search_times_df = pd.read_csv('linear_search_outputs.txt', sep=';')
linear_search_times_df.head()
```

```python
binary_search_times_df = pd.read_csv('binary_search_outputs.txt', sep=';')
binary_search_times_df.head()
```

# Análise de Tempo

```python
lin_n = np.array(linear_search_times_df['n'])
lin_t = np.array(linear_search_times_df['t'])
lin_i = np.array(linear_search_times_df.index)

plt.scatter(lin_i, lin_t, color='blue', label='linear')
plt.title('Tempo de execução da Busca Linear')
plt.xlabel('tamanho da lista [n]')
plt.ylabel('tempo de execução [s]')
```

```python
bin_n = np.array(binary_search_times_df['n'])
bin_t = np.array(binary_search_times_df['t'])
bin_i = np.array(binary_search_times_df.index)

plt.scatter(bin_i, bin_t, color='red')
plt.title('Tempo de execução da Busca Binária')
plt.xlabel('tamanho da lista [n]')
plt.ylabel('tempo de execução [s]')
```

```python
plt.scatter(lin_i, lin_t, color='blue', label='linear')
plt.scatter(bin_i, bin_t, color='red', label='binário')
plt.title('Comparação Busca Linear vs Busca Binária')
plt.xlabel('tamanho da lista [n]')
plt.ylabel('tempo de execução [s]')
plt.legend()
```

# Análise da quantidade de operações

```python
plt.scatter(lin_i, lin_n, color='blue', label='linear')
plt.title('Tempo de execução da Busca Linear')
plt.xlabel('tamanho da lista [n]')
plt.ylabel('quantidade de operações [n]')
```

```python
plt.scatter(bin_i, bin_n, color='red')
plt.plot(bin_i, np.log2(bin_i), label='curva log')

plt.title('Tempo de execução da Busca Binária')
plt.xlabel('tamanho da lista [n]')
plt.ylabel('quantidade de operações [n]')
plt.legend()
```
